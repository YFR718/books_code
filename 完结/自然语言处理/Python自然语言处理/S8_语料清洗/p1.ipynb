8-1 语料清洗

```python
text='？？？程序员是从事程序开发、维护的专业人员。%%##aba一般将程序员分为程序设计人员和程序编码人员，<body>但两者的78界限并不非常清楚，软件从业人员分为初级程序员、高级程序员、系统分析员和项目经理四大类。'
```

```python
import re

pattern = re.compile(r'[^\u4e00-\u9fa5]') # 去除数字、符号、字符等特殊字符
chinese_txt = re.sub(pattern,'',text)
print(chinese_txt)
```

8-2 df.fillna、dropna

```python
from numpy import nan
import pandas as pd

df1 = pd.DataFrame([[1,2,3],[nan,nan,2],[nan,nan,nan],[8,9,nan]])
print(df1)
df2 = df1.fillna(100)
print(df2)
df3 = df1.dropna()
print(df3)
```

8-3 Sklearn中Imputer类或SimpleImputer类

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=nan,strategy="mean")
df1[1] = imp.fit_transform(df1[[1]]) # 均值填充第1列Sklearn中Imputer类或SimpleImputer类
print(df1)
```

8-4 重复值清洗

```python
print(df1.drop_duplicates())
print(df1[0].drop_duplicates())
print(df1.drop_duplicates([1,2]))
```

```python
import missingno as msno
msno.matrix(df1,labels=True)
```
